```python
import findspark
findspark.init()
import sys
from pyspark import SparkContext,SparkConf
from time import time
```

```python
master = "local"
if len(sys.argv) == 2:
    master = sys.argv[1]
conf = SparkConf().setMaster(master).setAppName("rdd-set")
sc = SparkContext(conf = conf)
```

```python
file_path = "../data/kddcup.data_10_percent.gz"
raw_data = sc.textFile(file_path)
type(raw_data),raw_data.top(2),raw_data.count()
```

```python
normal_data = raw_data.filter(lambda x:"normal." in x)
normal_data.count()
```

```python
# 使用subtract算子获取不正常数据
attack_data = raw_data.subtract(normal_data)
attack_data.count()
```

```python
csv_data = raw_data.flatMap(lambda x:x.split(",")).distinct()
csv_data,csv_data.take(5)
```

```python
csv_data.collect()
```

```python
csv_data = raw_data.map(lambda x:x.split(","))
csv_data,csv_data.take(5)
```

```python
protocols = csv_data.map(lambda x:x[1]).distinct()
protocols.collect()
```

```python
services = csv_data.map(lambda x:x[2]).distinct()
type(services),services.collect(),services.count()
```

```python
# 笛卡尔积
products = protocols.cartesian(services).collect()
```
